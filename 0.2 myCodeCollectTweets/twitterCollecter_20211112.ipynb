# Collecting tweets

```python
# For dealing with json responses we receive from the API
import csv
import json
import time
import tweepy

import auth_function
import create_url_function
import connect_to_endpoint_function
import append_to_csv_function
import start_date_end_date
```

```python
#Inputs for tweets
bearer_token = auth_function.auth()
headers = auth_function.create_headers(bearer_token)
keyword = "(bmi)or(#bmi) -from:username lang:en"
start_list = start_date_end_date.start_list          
end_list = start_date_end_date.end_list
max_results = 50

#Total number of tweets we collected from the loop
total_tweets = 0

# Create file
csvFile = open("data.csv", "a", newline="", encoding='utf-8')
csvWriter = csv.writer(csvFile)

#Create headers for the data you want to save, in this example, we only want save these columns in our dataset
csvWriter.writerow(['author id', 'created_at', 'geo', 'id','lang', 'like_count', 'quote_count', 'reply_count','retweet_count','source','tweet'])
csvFile.close()
listCount=0

for i in range(0,len(start_list)):

    # Inputs
    count = 0 # Counting tweets per time period
    max_count = 100 # Max tweets per time period
    flag = True
    next_token = None
    
    # Check if flag is true
    while flag:
        # Check if max_count reached
        if count >= max_count:
            break
        print("-------------------")
        print("Token: ", next_token)
        url = create_url_function.create_url(keyword, start_list[i],end_list[i], max_results)
        json_response = connect_to_endpoint_function.connect_to_endpoint(url[0], headers, url[1], next_token)
        result_count = json_response['meta']['result_count']
        
        
        
        if 'next_token' in json_response['meta']:
            listCount+=1
            
            # Save the token to use for next call
            next_token = json_response['meta']['next_token']
            print("Next Token: ", next_token)
            if result_count is not None and result_count > 0 and next_token is not None:
                print("Start Date: ", start_list[i])
                print(json.dumps(json_response, indent=4, sort_keys=True))
                
                with open('original_data_'+str(start_list[i])+'_'+str(end_list[i])+'_'+str(listCount)+'.json', 'w') as f:
                    json.dump(json_response, f)
                
                count += result_count
                total_tweets += result_count
                print("Total # of Tweets added: ", total_tweets)
                print("-------------------")
                time.sleep(5)                
        # If no next token exists
        else:
            if result_count is not None and result_count > 0:
                listCount+=1
                
                print("-------------------")
                print("Start Date: ", start_list[i])
                #append_to_csv_function.append_data_to_csv(json_response, "data.csv")
                #print(json.dumps(json_response, indent=4, sort_keys=True))
                with open(with open('original_data_'+str(start_list[i])+'_'+str(end_list[i])+'_'+str(listCount)+'.json', 'w') as f:
                    json.dump(json_response, f)
                
                count += result_count
                total_tweets += result_count
                print("Total # of Tweets added: ", total_tweets)
                print("-------------------")
                time.sleep(5)
            
            #Since this is the final request, turn flag to false to move to the next time period.
            flag = False
            next_token = None
        time.sleep(5)
print("Total number of results: ", total_tweets)
```
